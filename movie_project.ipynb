```python
import ast
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i["name"])

  return l
```

```python
data["genres"] = data["genres"].apply(convert)
```

```python
data
```

```python
data["keywords"][0]
```

```python
data["keywords"] = data["keywords"].apply(convert)
data
```

```python
data["cast"][0]
```

```python
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter < 3:
      l.append(i["name"])
      counter += 1
    else:
      break

  return l
```

```python
data["cast"] = data["cast"].apply(convert3)
data["cast"][0]
```

```python
data
```

```python
data["crew"][0]
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      L.append(i["name"])

  return L
```

```python
data["crew"] = data["crew"].apply(fetch_director)
data["crew"][0]
```

```python
data["overview"] = data["overview"].apply(lambda x : x.split())
```

```python
data["overview"]
```

```python
data["genres"] = data["genres"].apply(lambda x : [i.replace(" ", "") for i in x])
data["keywords"] = data["keywords"].apply(lambda x : [i.replace(" ", "") for i in x])
data["cast"] = data["cast"].apply(lambda x : [i.replace(" ", "") for i in x])
data["crew"] = data["crew"].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
data.head()
```

```python
data["tags"] = data["overview"] + data["genres"] + data["keywords"] + data["cast"] + data["crew"]
data
```

```python
print(data["tags"][0])
```

```python
new_df = data[["movie_id", "title", "tags"]]
new_df
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x : " ".join(x))
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x : x.lower())
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')

from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize, sent_tokenize
```

```python
len(new_df["tags"][0])
```

```python
def lem(text):
  y = []
  for i in text.split():
    y.append(WordNetLemmatizer().lemmatize(i))
  return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(lem)
```

```python
len(new_df["tags"][0])
```

```python
new_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfid_matrix = TfidfVectorizer(max_features = 5000, stop_words = "english").fit_transform(new_df["tags"]).toarray()
tfid_matrix
```

```python
tfid_matrix[0]
```

```python
tfid_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(tfid_matrix, tfid_matrix)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
new_df[new_df["title"] == "Spectre"].index[0]
```

```python
def recommende(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  distance = similarity[movie_index]
  movie_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x : x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]]["title"])
```

```python
recommende("Avatar")
```
